```python
import os

def count_files_in_directories(root_dir):
    # Dictionary to store the count of files in each directory
    dir_file_counts = {}

    # Iterate over all directories and subdirectories
    for dirpath, dirnames, filenames in os.walk(root_dir):
        # Count the files in the current directory
        file_count = len(filenames)
        
        # Store the count in the dictionary
        dir_file_counts[dirpath] = file_count

    return dir_file_counts

    # Specify the root directory
root_directory = r'C:\Users\tomce\OneDrive - UKIM, FINKI\Desktop\Fakultet 3ta Godina\2 Sesti Semestar\0 DATA SCIENCE SEMINARSKA\1 Starting Over\products_with_categories'

    # Get the file counts
file_counts = count_files_in_directories(root_directory)

    # Sort the results by file count in descending order
sorted_file_counts = sorted(file_counts.items(), key=lambda item: item[1], reverse=True)

    # Print the sorted results
for dirpath, count in sorted_file_counts:
    print(f"Directory: {dirpath} has {count} files")
```
