```python
import os
import shutil

# Directories
tvs_new_template_dir = r'C:\Users\tomce\OneDrive - UKIM, FINKI\Desktop\Fakultet 3ta Godina\2 Sesti Semestar\0 DATA SCIENCE SEMINARSKA\1 Starting Over\products_categories_fixed\all_extracted_with_lists_dicts'
products_with_categories_all_dir = r'C:\Users\tomce\OneDrive - UKIM, FINKI\Desktop\Fakultet 3ta Godina\2 Sesti Semestar\0 DATA SCIENCE SEMINARSKA\1 Starting Over\products_categories_fixed\products_with_categories_all'
destination_dir = r'C:\Users\tomce\OneDrive - UKIM, FINKI\Desktop\Fakultet 3ta Godina\2 Sesti Semestar\0 DATA SCIENCE SEMINARSKA\1 Starting Over\products_categories_fixed\moved_files'

# Create the destination directory if it doesn't exist
if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)

# Step 1: List all JSON files in the tvs_new_template directory and remove the "processed_" prefix
tvs_new_template_files = os.listdir(tvs_new_template_dir)
processed_files = {f.replace('processed_', '') for f in tvs_new_template_files if f.endswith('.json')}

# Step 2: List all JSON files in the products_with_categories_all directory
products_with_categories_all_files = os.listdir(products_with_categories_all_dir)
products_files_set = {f for f in products_with_categories_all_files if f.endswith('.json')}

# Step 3: Identify the files in products_with_categories_all that should be moved
files_to_move = products_files_set.intersection(processed_files)

# Step 4: Move the identified files from the products_with_categories_all directory to the destination directory
for file_name in files_to_move:
    src_path = os.path.join(products_with_categories_all_dir, file_name)
    dest_path = os.path.join(destination_dir, file_name)
    shutil.move(src_path, dest_path)
    print(f"Moved: {src_path} to {dest_path}")

print("Files moved successfully.")
```
