#### This notebook was used to move the JSON files of each directory from the "products_with_categories" folder to the "products_with_categories_all" folder. This was intended for the universal extractor, which DID NOT WORK AS INTENDED.

```python
import os
import shutil

def move_json_files(src_dir, dst_dir):
    # Ensure the destination directory exists
    if not os.path.exists(dst_dir):
        os.makedirs(dst_dir)

    # Walk through all directories and subdirectories
    for root, dirs, files in os.walk(src_dir):
        for file in files:
            if file.endswith('.json'):
                # Construct full file path
                file_path = os.path.join(root, file)
                # Move file to destination directory
                shutil.copy(file_path, os.path.join(dst_dir, file))
                print(f'Moved: {file_path} to {dst_dir}')

# Specify source and destination directories
source_directory = r'C:\Users\tomce\OneDrive - UKIM, FINKI\Desktop\Fakultet 3ta Godina\2 Sesti Semestar\0 DATA SCIENCE SEMINARSKA\1 Starting Over\products_categories_fixed\products_with_categories'
destination_directory = r'C:\Users\tomce\OneDrive - UKIM, FINKI\Desktop\Fakultet 3ta Godina\2 Sesti Semestar\0 DATA SCIENCE SEMINARSKA\1 Starting Over\products_categories_fixed\products_with_categories_all'

# Call the function
move_json_files(source_directory, destination_directory)
```
